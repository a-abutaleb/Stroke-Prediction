data seems to have same values over a range, so we actually do not have a random data. So need to shuffle it to make it random

```python
train_data = train_data.sample(frac=1).reset_index(drop=True)
train_data.to_csv('/content/drive/MyDrive/Logistic Regression on Stroke prediction/StrokeDataTrain.csv', index=False)
# we put the file address to rewrite the shuffled file into our previous file
train_data.head(20)
```

#for single column to encode

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
column_to_encode = ['ever_married']
encoder.fit(train_data[column_to_encode])
encoded_column = encoder.transform(train_data[column_to_encode])
train_data['ever_married']=encoded_column # this is replacing back the encoded column to original data
```

```python
train_data.head(5)
```

#for multiple columns to encode

```python
columns_to_encode = ['gender', 'work_type', 'Residence_type', 'smoking_status']
for column in columns_to_encode:
 encoded_columns = encoder.fit_transform(train_data[column])
 train_data[column] = encoded_columns
```

```python
train_data.head(5)
```

```python
train_data.isnull().sum()
```

```python
train_data['bmi'].fillna(train_data['bmi'].mean(), inplace=True)
```

```python
train_data.isnull().sum()
```

```python
X_train = train_data[['id','gender', 'age', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'avg_glucose_level', 'bmi', 'smoking_status']]
Y_train = train_data['stroke']
log_reg = linear_model.LogisticRegression()
log_model = log_reg.fit(X_train,Y_train)
test_data = pd.read_csv("/content/drive/MyDrive/Logistic Regression on Stroke prediction/test.csv")
test_data.head(5)
```

```python
columns_to_encode = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
for column in columns_to_encode:
 encoded_columns = encoder.fit_transform(test_data[column])
 test_data[column] = encoded_columns
```
